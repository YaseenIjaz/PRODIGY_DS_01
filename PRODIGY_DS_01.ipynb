```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("world_population.csv")
df
```

```python
df.info()
```

```python
df.columns
```

```python
# Sort the data by Density (per km²) for better visualization
df_sorted = df.sort_values(by='Density (per km²)', ascending=False).head(20)  # Top 20 countries by density

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(df_sorted['Country/Territory'], df_sorted['Density (per km²)'], color='skyblue')
plt.xlabel('Country/Territory')
plt.ylabel('Population Density (per km²)')
plt.title('Top 20 Countries by Population Density')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
